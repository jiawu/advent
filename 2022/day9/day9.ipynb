```python
import pandas as pd
import numpy as np
```

```python
with open('input.txt') as fin:
    instructions=fin.readlines()
```

```python
instructions = [x.rstrip().split(' ') for x in instructions]
```

```python
np.add((1,0), (1,0))
```

```python
move_it={'U': (1,0),
     'D': (-1,0),
     'L': (0,-1),
     'R': (0,1)}

possible_tail_moves = [(1,1), (0,1), (1,0), (1,-1), (-1,1), (-1,-1), (0,-1), (-1,0)] 
```

```python
def calc_dist(point1, point2):
    dist = np.linalg.norm(point1 - point2)
    return(dist)
```

```python
current_head_coordinate = (0,0)
current_tail_coordinate = (0,0)

head_coordinates = [np.array([0,0])]
tail_coordinates = [np.array([0,0])]
for move in instructions:
    for x in range(0,int(move[1])):
        current_head_coordinate = np.add(current_head_coordinate, move_it[move[0]])
        head_coordinates.append(current_head_coordinate)
        current_dist = calc_dist(current_tail_coordinate, current_head_coordinate)
        if current_dist < 2:
            continue
        else:
            tail_distances = []
            for tm in possible_tail_moves:
                tail_distances.append(calc_dist(current_head_coordinate, np.add(current_tail_coordinate, tm)))
            pos_tail_dist = dict(zip(possible_tail_moves, tail_distances))
            selected_move=min(pos_tail_dist, key=pos_tail_dist.get)
            current_tail_coordinate = np.add(current_tail_coordinate, selected_move)
            tail_coordinates.append(current_tail_coordinate)
```

```python
tail_df=pd.DataFrame(tail_coordinates)
```

```python
tail_df.drop_duplicates(subset=[0,1]
```

```python
all_coordinates = [np.array([0,0]) for x in range(0,10)]
all_coordinates_history = [list([np.array([0,0])]) for x in range(0,10)]
for move in instructions:
    for x in range(0,int(move[1])):
        # current_head_coordinate
        all_coordinates[0] = np.add(all_coordinates[0], move_it[move[0]])
        all_coordinates_history[0].append(all_coordinates[0])
        for n in range(0,9):            
            current_dist = calc_dist(all_coordinates[n], all_coordinates[n+1])
            if current_dist < 2:
                all_coordinates_history[n+1].append(all_coordinates[n+1])
                continue
            else:
                tail_distances = []
                for tm in possible_tail_moves:
                    tail_distances.append(calc_dist(all_coordinates[n], np.add(all_coordinates[n+1], tm)))
                pos_tail_dist = dict(zip(possible_tail_moves, tail_distances))
                selected_move=min(pos_tail_dist, key=pos_tail_dist.get)
                all_coordinates[n+1] = np.add(all_coordinates[n+1], selected_move)
                all_coordinates_history[n+1].append(all_coordinates[n+1])
```
