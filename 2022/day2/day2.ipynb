```python
import pandas as pd
import numpy as np
```

```python
strat=pd.read_csv('input.txt', header=None, delimiter=' ')
```

```python
strat.columns = ['player1', 'player2']
```

```python
strat['player1_plays'] = strat['player1'].replace({'A':'rock', 'B':'paper','C':'scissors'})
strat['player2_plays'] = strat['player2'].replace({'X':'rock', 'Y':'paper','Z':'scissors'})
```

```python
strat
```

```python
strat['innate']= strat['player2_plays'].replace({'rock':1,'paper':2,'scissors':3})

conditions = [
    (strat['player1_plays'] == strat['player2_plays']),
    (strat['player2_plays'] == 'rock') & (strat['player1_plays'] == 'scissors'),
    (strat['player2_plays'] == 'paper') & (strat['player1_plays'] == 'rock'),
    (strat['player2_plays'] == 'scissors') & (strat['player1_plays'] == 'paper')]
score = [3,6,6,6]
strat['score'] = np.select(conditions,score,default=0)
```

```python
strat['total_score'] = strat['innate'] + strat['score']
```

```python
strat['total_score'].sum()
```

```python
strat['part2_score'] = strat['player2'].replace({'X':0, 'Y':3,'Z':6})
strat
```

```python
conditions = [
    (strat['player1_plays']== 'rock') & (strat['part2_score'] == 6),
    (strat['player1_plays']== 'rock') & (strat['part2_score'] == 3),
    (strat['player1_plays']== 'rock') & (strat['part2_score'] == 0),
    (strat['player1_plays']== 'paper') & (strat['part2_score'] == 6),
    (strat['player1_plays']== 'paper') & (strat['part2_score'] == 3),
    (strat['player1_plays']== 'paper') & (strat['part2_score'] == 0),
    (strat['player1_plays']== 'scissors') & (strat['part2_score'] == 6),
    (strat['player1_plays']== 'scissors') & (strat['part2_score'] == 3),
    (strat['player1_plays']== 'scissors') & (strat['part2_score'] == 0)]
score = [2,1,3, 3,2,1, 1,3,2]
strat['part2_innate_score'] = np.select(conditions,score)
```
