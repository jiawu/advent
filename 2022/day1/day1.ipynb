```python
import pandas as pd
```

```python
with open('input.txt') as fin:
    calories=fin.readlines()
```

```python
all_elves = []
elf = []
for line in calories:
    new_line=line.strip()
    if len(new_line) == 0:
        all_elves.append(elf)
        elf = []
    else:
        elf.append(int(new_line))

all_elves.append(elf)
```

```python
all_elves=pd.DataFrame(all_elves)
```

```python
all_elves.sum(axis=1).max()
```
