```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('input.txt', header=None)
```

```python
with open('input.txt') as fin:
    instruct=fin.readlines()
instruct = [x.rstrip() for x in instruct]
```

```python
instruct
cycle_num = 1
cycle_num2 = 1
cycle_track = []
cycle_track2 = []
x = 1
x_track = []
for ins in instruct:
    cycle_num2 = cycle_num
    if ins == 'noop':
        cycle_num += 1
    elif ins.startswith('addx'):
        cycle_num += 2
        add = int(ins.split(' ')[1])
        x += add
    cycle_track.append(cycle_num)
    cycle_track2.append(cycle_num2)
    x_track.append(x)
```

```python
df['cycle_start'] = cycle_track2
df['cycle_end'] = cycle_track
df['x_track'] = x_track
```

```python
df
```

```python
real_track = pd.DataFrame()
```

```python
real_track['cycle_num'] = [i for i in range(1,df['cycle_end'].tolist()[-1]+1)]
```

```python
real_track = real_track.merge(df, how='left', left_on='cycle_num', right_on='cycle_end')
```

```python
sprite_pos = real_track['x_track'].tolist()
```

```python
output = ""
for ind, pix in enumerate(range(0, 240)):
    if ind%40 == 0:
        output = output + '\n'
    
    sp = sprite_pos[ind]
    current_sprite = [sp-1, sp, sp+1]
    conv_pix = pix % 40
    
    if conv_pix in current_sprite:
        output = output + "#"
    else:
        output= output+'.'
```
