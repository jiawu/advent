```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('input.txt', header=None)
```

```python
df.columns = ['agg']
```

```python
df['total_size'] = df['agg'].apply(lambda x: len(x))
```

```python
df['total_size'] = df['total_size'].astype(int)
```

```python
compart1 = []
compart2 = []
compart1_set = []
compart2_set = []
intersection = []
for num,row in df.iterrows():
    compart1.append(row['agg'][0:int(row['total_size']/2)])
    compart2.append(row['agg'][int(row['total_size']/2):])
    compart1_set.append(set(list(row['agg'][0:int(row['total_size']/2)])))
    compart2_set.append(set(list(row['agg'][int(row['total_size']/2):])))
    intersection.append(compart1_set[num].intersection(compart2_set[num]))
```

```python
df['compart1'] = compart1
df['compart2'] = compart2
df['compart1_set'] = compart1_set
df['compart2_set'] = compart2_set
df['intersection'] = intersection
```

```python
# lowercase = -26
# uppercase = -64+27= -38
ord('Z')-38
```

```python
df['intersection_isupper'] = df['intersection'].apply(lambda x: list(x)[0].isupper())
```

```python
df['intersection_isupper']
```

```python
ord('W')-38
```

```python
df['intersection_ord'] = df['intersection'].apply(lambda x: ord(list(x)[0]))
```

```python
df['priority']=np.where(df['intersection_isupper'], df['intersection_ord']-38, df['intersection_ord'] - 96)
```

```python
inds=[x for x in range(0,300)]
res =[list(x) for x in np.split(np.array(inds), 100) if x.size != 0]
```

```python
badges = []
for group in res:
    elf_clade = []
    for elf in group:
        elf_clade.append(set(df['agg'].iloc[elf]))
    badges.append(elf_clade[0].intersection(elf_clade[1]).intersection(elf_clade[2]))
```

```python
badges=[list(x)[0] for x in badges]
```
