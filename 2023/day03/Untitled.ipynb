```python
import pandas as pd
import re
import pdb
```

```python
with open("input.txt") as f:
    data = f.read().strip()
```

```python
parsed = data.split('\n')
```

```python
y = len(parsed)
x = len(parsed[0])
```

```python
symbol_ind = []
for cur_y in range(0, y):
    for cur_x in range(0 ,x):
        current_char = parsed[cur_y][cur_x]
        if bool(re.match(r"^[^a-zA-Z0-9.]+$", current_char)):
                symbol_ind.append((cur_x,cur_y))
```

```python
print(parsed[symbol_ind[0][1]][symbol_ind[0][0]])
```

```python
# for each symbol_ind, check for numbers around it
# if loc is a number, check for numbers behind or infront of it
```

```python
direction = {'N': (-1,0),
 'NW': (-1,-1),
 'NE': (-1,1),
 'E':(0,1),
 'W':(0,-1),
 'S':(1,0),
 'SE':(1,1),
 'SW':(1,-1)}
```

```python
all_parts = []
for cur_sym in symbol_ind:
    current_char = parsed[cur_sym[1]][cur_sym[0]]
    print(current_char)
    print(cur_sym)
    all_dir_sym = {'N':parsed[cur_sym[1]-1][cur_sym[0]],
    'NW':parsed[cur_sym[1]-1][cur_sym[0]-1],
    'NE':parsed[cur_sym[1]-1][cur_sym[0]+1],
    'E':parsed[cur_sym[1]][cur_sym[0]+1],
    'W': parsed[cur_sym[1]][cur_sym[0]-1],
    'S':parsed[cur_sym[1]+1][cur_sym[0]],
    'SE':parsed[cur_sym[1]+1][cur_sym[0]+1],
    'SW':parsed[cur_sym[1]+1][cur_sym[0]-1]}    
    #all_dir = [N, NW, NE, E, W, S, SE, SW]
    
    for k,v in all_dir_sym.items():
        if v.isdigit():
            part_num = v
            xi, yi = direction[k]
            num_ind = [(cur_sym[1] + xi, cur_sym[0] + yi)]
            for i in range(1,4):
                if cur_sym[0] + yi + i < x:
                    next_let = parsed[cur_sym[1] + xi][cur_sym[0] + yi + i]
                    if next_let.isdigit():
                        part_num = part_num + next_let
                        num_ind.append((cur_sym[1] + xi, cur_sym[0] + yi + i))
                    else:
                        break
            for i in range(1,4):
                prev_let = parsed[cur_sym[1] + xi][cur_sym[0] + yi - i]
                if prev_let.isdigit():
                    part_num = prev_let + part_num 
                    num_ind.append((cur_sym[1] + xi, cur_sym[0] + yi - i))
                    
                else:
                    break
            pre = set(prev_ind)
            ne = set(num_ind)
            if (pre != ne) and ~(pre <= ne) and ~(ne <= pre):
                all_parts.append(part_num)
            prev_ind = num_ind
```

```python
sum([int(x) for x in all_parts])
```

```python
all_parts = []
for cur_sym in symbol_ind:
    current_char = parsed[cur_sym[1]][cur_sym[0]]
    print(current_char)
    print(cur_sym)
    all_dir_sym = {'N':parsed[cur_sym[1]-1][cur_sym[0]],
    'NW':parsed[cur_sym[1]-1][cur_sym[0]-1],
    'NE':parsed[cur_sym[1]-1][cur_sym[0]+1],
    'E':parsed[cur_sym[1]][cur_sym[0]+1],
    'W': parsed[cur_sym[1]][cur_sym[0]-1],
    'S':parsed[cur_sym[1]+1][cur_sym[0]],
    'SE':parsed[cur_sym[1]+1][cur_sym[0]+1],
    'SW':parsed[cur_sym[1]+1][cur_sym[0]-1]}    
    #all_dir = [N, NW, NE, E, W, S, SE, SW]
    part_count = 0
    cur_part = []
    for k,v in all_dir_sym.items():
        if v.isdigit():
            part_num = v
            xi, yi = direction[k]
            num_ind = [(cur_sym[1] + xi, cur_sym[0] + yi)]
            for i in range(1,4):
                if cur_sym[0] + yi + i < x:
                    next_let = parsed[cur_sym[1] + xi][cur_sym[0] + yi + i]
                    if next_let.isdigit():
                        part_num = part_num + next_let
                        num_ind.append((cur_sym[1] + xi, cur_sym[0] + yi + i))
                    else:
                        break
            for i in range(1,4):
                prev_let = parsed[cur_sym[1] + xi][cur_sym[0] + yi - i]
                if prev_let.isdigit():
                    part_num = prev_let + part_num 
                    num_ind.append((cur_sym[1] + xi, cur_sym[0] + yi - i))

                else:
                    break
            pre = set(prev_ind)
            ne = set(num_ind)
            if (pre != ne) and ~(pre <= ne) and ~(ne <= pre):
                part_count += 1
                cur_part.append(part_num)
            prev_ind = num_ind
    if part_count == 2:
        all_parts.append(int(cur_part[0])* int(cur_part[1]))
```

```python
sum([int(x) for x in all_parts])
```
